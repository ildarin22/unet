```python
from keras.preprocessing.image import ImageDataGenerator
from model import *
```

```python
data_gen_args = dict(rotation_range=0.2,
                    width_shift_range=0.05,
                    height_shift_range=0.05,
                    shear_range=0.05,
                    zoom_range=0.05,
                    horizontal_flip=True,
                    fill_mode='nearest')
```

```python
image_datagen = ImageDataGenerator(**data_gen_args)
image_generator = image_datagen.flow_from_directory(
    'data/cats/train',
    classes = ['image'],
    class_mode = None,
    color_mode = 'rgb',
        target_size = (256,256),
        batch_size = 62,
        save_to_dir = 'data/cats/train/aug',
        save_prefix  = 'image',
        seed = 1
)
```

```python
model = unet()
model_checkpoint = ModelCheckpoint('unet_cats.hdf5', monitor='loss',verbose=1, save_best_only=True)
model.fit_generator(image_generator,steps_per_epoch=2000,epochs=5,callbacks=[model_checkpoint])
```
